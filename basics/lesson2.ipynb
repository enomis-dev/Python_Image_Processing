# Learn how to add figures or text to blank images

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a blank image
blank = np.zeros((500,500,3), dtype='uint8')
plt.imshow(blank)
plt.title('Blank image')
```

```python
# 1. Color a portion of the image in blue
blank[200:300, 300:400] = 0,0,255
plt.imshow(blank)
plt.title('Blue image')
```

```python
# 2. Draw a Rectangle
cv.rectangle(blank, (0,0), (250,250), (0,255,0), thickness=cv.FILLED) # cv.Filled to add a full color
plt.imshow(blank)
plt.title('Rectangle')
```

```python
# 3. Draw a circle
# we are giving the coordinates of the image center
cv.circle(blank, (blank.shape[1]//2,blank.shape[0]//2), 40, (0,0,255), thickness=-1)
plt.imshow(blank)
plt.title('Circle')
```

```python
# 4. Draw a line
cv.line(blank, (0,0), (blank.shape[1]//2, blank.shape[0]//2), (255,255,255), thickness=3)
plt.imshow(blank)
plt.title('Line')
```

```python
# 5. Write text
cv.putText(blank, 'Hello', (225,225), cv.FONT_HERSHEY_TRIPLEX, 1.0, (255,0,0), thickness=2)
plt.imshow(blank)
plt.title('Text')
```
