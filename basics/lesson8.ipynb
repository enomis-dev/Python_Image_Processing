# Objective: learn how to apply gradients to an image

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read the image on the disk
source_path = Path().absolute().parent
image_path = os.path.join(source_path, 'Data', 'dog.jpg')
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('RGB image')
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

# Laplacian
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
plt.imshow(lap, cmap='Greys')
plt.title('Laplacian image')
```

```python
# Sobel
sobelx = cv.Sobel(gray, cv.CV_64F,1,0)
sobely = cv.Sobel(gray, cv.CV_64F,0,1)
combined_sobel = cv.bitwise_or(sobelx, sobely)


fig, axs = plt.subplots(1, 3, figsize=(11, 10))

# they will be displayed in grayscale, actually this is the intensity of that color
# lighter -->  bigger concentration
axs[0].imshow(sobelx, cmap='Greys')
axs[0].set_title('Sobel x')
axs[1].imshow(sobely, cmap='Greys')
axs[1].set_title('Sobel y')
axs[2].imshow(combined_sobel, cmap='Greys')
axs[2].set_title('Combined Sobel')
```

```python
# Canny, it a multistage algorithm and it uses the Sobel
# method to compute the gradient of the images in one of its steps
canny = cv.Canny(gray, 150, 175)
plt.imshow(canny, cmap='Greys')
plt.title('Canny image')
```
