# Objective: Learn how to split an image in its 3 colors

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read the image on the disk
source_path = Path().absolute().parent
image_path = os.path.join(source_path, 'Data', 'dog.jpg')
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('RGB image')
```

```python
# create a blank image
blank = np.zeros(img.shape[:2], dtype='uint8')
r,g,b = cv.split(img)

fig, axs = plt.subplots(1, 3, figsize=(12, 8))

# they will be displayed in grayscale, actually this is the intensity of that color
# lighter -->  bigger concentration
axs[0].imshow(b, cmap='Greys')
axs[0].set_title('Blue image')
axs[1].imshow(g, cmap='Greys')
axs[1].set_title('Green image')
axs[2].imshow(r, cmap='Greys')
axs[2].set_title('Red image')
```

```python
# merge
merged = cv.merge([b,g,r])
cv.imshow('Merge',merged)

# If we want to display the color with have to convert the image to a 3 channel image
# Display the colors
blue = cv.merge([b,blank,blank])
green = cv.merge([blank,g,blank])
red = cv.merge([blank,blank,r])

fig, axs = plt.subplots(1, 3, figsize=(12, 8))

axs[0].imshow(blue, cmap='Blues')
axs[0].set_title('Blue image')
axs[1].imshow(green, cmap='Greens')
axs[1].set_title('Green image')
axs[2].imshow(red, cmap='Reds')
axs[2].set_title('Red image')
```
