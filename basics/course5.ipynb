# Objective: Learn how to convert to other color spaces

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read the image on the disk
source_path = Path().absolute().parent
image_path = os.path.join(source_path, 'Data', 'dog.jpg')
img = cv.imread(image_path)

plt.imshow(img)
plt.title('BGR image')
```

```python
# BGR to GrayScale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Gray image')
```

```python
# BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.title('HSV image')
```

```python
# BGR to L*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
plt.imshow(lab)
plt.title('LAB image')
```

```python
# BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title('RGB image')
```
