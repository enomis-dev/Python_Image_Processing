# Objective: Blurring an image

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read the image on the disk
source_path = Path().absolute().parent
image_path = os.path.join(source_path, 'Data', 'dog.jpg')
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('RGB image')
```

```python
# Averaging blur (we define a kernel size and we blur the image)
average = cv.blur(img, (7,7))

plt.imshow(average)
plt.title('Average Blur')
```

```python
# Gaussian blur, it is more natural as compared to the gaussian one
gauss = cv.GaussianBlur(img, (7,7), 0)
plt.imshow(gauss)
plt.title('Gaussian Blur')
```

```python
# Median Blur, is like average
# but it take the median
```

```python
# Median Blur, is like average
# but it take the median

median = cv.medianBlur(img,7)
plt.imshow(median)
plt.title('Median Blur')
```

```python
# Bilateral blurring
# it is the most effective, traditional blurring methods
# don't care about edges. Bilateral takes care
bilateral = cv.bilateralFilter(img, 5, 15, 15)
plt.imshow(bilateral)
plt.title('Bilateral')
```
