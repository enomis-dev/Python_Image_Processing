# Objective: Learn how to apply a mask to an image

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read the image on the disk
source_path = Path().absolute().parent
image_path = os.path.join(source_path, 'Data', 'dog.jpg')
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('RGB image')
```

```python
# we create a blank image taking the image shape and then we add a mask
blank = np.zeros(img.shape[:2], dtype='uint8')
plt.imshow(blank, cmap='Greys')
plt.title('Blank image')
```

```python
mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
plt.imshow(mask, cmap='Greys')
plt.title('Mask')
```

```python
masked = cv.bitwise_and(img, img, mask=mask)
plt.imshow(masked, cmap='Greys')
plt.title('Masked Image')
```
