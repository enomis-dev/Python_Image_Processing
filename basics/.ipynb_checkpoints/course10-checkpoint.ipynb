# Objective: Learn how to do bitwise operations

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
blank = np.zeros((400,400), dtype='uint8')
rectangle = cv.rectangle(blank.copy(),(30,30),(370,370),255,-1)
circle = cv.circle(blank.copy(),(200,200),200,255,-1)
```

```python
plt.imshow(rectangle, cmap='Greys')
plt.title('Rectangle')
```

```python
plt.imshow(circle, cmap='Greys')
plt.title('Circle')
```

```python
# bitwise AND --> intersecting regions
bitwise_and = cv.bitwise_and(rectangle, circle)
plt.imshow(bitwise_and, cmap='Greys')
plt.title('Bitwise AND')
```

```python
# bitwise OR --> non intersecting and intersecting region
bitwise_or = cv.bitwise_or(rectangle, circle)
plt.imshow(bitwise_or, cmap='Greys')
plt.title('Bitwise OR')
```

```python
# bitwise XOR --> non-intersecting regions (OR - AND)
bitwise_xor = cv.bitwise_xor(rectangle, circle)
plt.imshow(bitwise_xor, cmap='Greys')
plt.title('Bitwise XOR')
```

```python
# invert white and black
bitwise_not = cv.bitwise_not(rectangle)
plt.imshow(bitwise_not, cmap='Greys')
plt.title('Bitwise NOT')
```
