# Objective: Learn how to train the haar cascade model for face recognition
The trained model is saved in a .yml file and used to recognize an unseen photo

```python
# import libraries
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# create a list with names
people = []
source_path = Path().absolute().parent
DIR = os.path.join(source_path, 'Data', 'FACES', 'train')
for i in os.listdir(DIR):
    people.append(i)

features = []
labels = []

# we read the classifier model
model_path = os.path.join(source_path, 'detection', 'haar_face.xml')
haar_cascade = cv.CascadeClassifier(model_path)

def create_train():
    for person in people:
        path = os.path.join(DIR, person)
        label = people.index(person)

        for img in os.listdir(path):
            img_path = os.path.join(path, img)

            img_array = cv.imread(img_path)
            gray = cv.cvtColor(img_array, cv.COLOR_BGR2GRAY)

            faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1,
                                                       minNeighbors=4)

            for (x,y,w,h) in faces_rect:
                faces_roi = gray[y:y+h, x:x+w]
                features.append(faces_roi)
                labels.append(label)

create_train()

print(f'Length of the features = {len(features)}')
print(f'Length of the labels = {len(labels)}')
print('Training done---------')

features = np.array(features, dtype='object')
labels = np.array(labels)


# Create a face recognizer object
face_recognizer = cv.face.LBPHFaceRecognizer_create()

# Train the Recognizer on the features list and labels list
face_recognizer.train(features, labels)
```

```python
# we save the recognizer to a yml file
dst_path = os.path.join(source_path, 'detection', 'face_trained.yml')
face_recognizer.save(dst_path)
dst_path_features = os.path.join(source_path, 'detection', 'features.npy')
np.save(dst_path_features, features)
dst_path_labels = os.path.join(source_path, 'detection', 'labels.npy')
np.save(dst_path_labels, labels)
```

```python
haar_cascade = cv.CascadeClassifier(model_path)

#features = np.load(dst_path_features)
#labels = np.load(dst_path_labels)

face_recognizer = cv.face.LBPHFaceRecognizer_create()
# read the pre-trained model
face_recognizer.read(dst_path)
img = cv.imread(os.path.join(source_path, 'Data', 'Faces', 'val', 'elton_john', '1.jpg'))
#img = cv.resize(img, (500,500), interpolation=cv.INTER_CUBIC)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.title('Person')

# Detect the face in the image
faces_rect = haar_cascade.detectMultiScale(gray, 1.1, 4)

for (x,y,w,h) in faces_rect:
    faces_roi = gray[y:y+h,x:x+h]

    label, confidence = face_recognizer.predict(faces_roi)
    print(f'label = {people[label]}  with a confidence of {confidence}')

    cv.putText(img, str(people[label]), (20,20), cv.FONT_HERSHEY_COMPLEX, 1.0, (0,255,0),
               thickness =2)
    cv.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)
```
